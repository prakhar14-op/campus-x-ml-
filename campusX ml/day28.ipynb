```python
import pandas as pd 
import numpy as np
```

```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder
```

step 1: train test split

```python
df=pd.read_csv('covid_toy.csv')
df.sample(5)
```

```python

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(df.drop(columns=['has_covid']),df['has_covid'],test_size=0.2)
x_train.shape,x_test.shape
```

here fever is a numerical column but have missing values so ham transfromation lagaynge ispe SimpleImputer ye missing value ki jagha mean fill kar dega


```python
si=SimpleImputer()
x_train_fever=si.fit_transform(x_train[['fever']])

x_test_fever=si.transform(x_test[['fever']])

x_train_fever
```

```python
# ordinalencoding-->cough
oe=OrdinalEncoder()
x_train_cough=oe.fit_transform(x_train[['cough']])

x_test_cough=oe.transform(x_test[['cough']])
x_train_cough
```

```python
# OneHotEncoder-->city,gender
ohe=OneHotEncoder(drop='first',sparse_output=False)
x_train_gender_city=ohe.fit_transform(x_train[['gender','city']])

x_test_gender_city=ohe.transform(x_test[['gender','city']])
x_train_gender_city #first column is of gender then other three is for city 
x_test_gender_city.shape
```

```python
# extracting age 
x_train_age=x_train.drop(columns=['gender','city','fever','cough']).values

x_test_age=x_test.drop(columns=['gender','city','fever','cough']).values

x_train_age,x_test_age.shape
```

```python
print(x_test_cough.shape)
print(x_test_age.shape)
print(x_test_fever.shape)
print(x_test_gender_city.shape)

print(x_train_age.shape)
print(x_train_cough.shape)
print(x_train_gender_city.shape)
print(x_train_fever.shape)
```

```python
# concatinating all the columns
x_train_transformed=np.concatenate((x_train_age,x_train_cough,x_train_gender_city,x_train_fever),axis=1)

x_test_transformed=np.concatenate((x_test_age,x_test_cough,x_test_gender_city,x_test_fever),axis=1)

x_train_transformed
```

method 2: using column transform

```python
from sklearn.compose import ColumnTransformer
transformer=ColumnTransformer(transformers=[
    ('tnf1',SimpleImputer(),['fever']),
    ('tnf2',OrdinalEncoder(categories=[['Mild','Strong']]),['cough']),
    ('tnf3',OneHotEncoder(drop='first',sparse_output=False),['city','gender']),
],remainder='passthrough')
```

```python
transformer.fit_transform(x_train)
```
