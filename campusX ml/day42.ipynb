```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('outlier_placement.csv')
df.sample(5)
```

step 2: ploting ddistpot and boxplot of each column

```python
plt.figure(figsize=(16,8))
plt.subplot(2,2,1)
sns.distplot(df['cgpa'])

plt.subplot(2,2,2)
sns.boxplot(df['cgpa'])

plt.subplot(2,2,3)
sns.distplot(df['placement_exam_marks'])

plt.subplot(2,2,4)
sns.boxplot(df['placement_exam_marks'])
plt.show()
```

step 3: finding the heighst allowed value and lowest allowed value 

```python
print('heighest allowed',df['cgpa'].mean()+3*df['cgpa'].std())
print('lowest allowed ',df['cgpa'].mean()-3*df['cgpa'].std())
```

step 4:trimming the outliers and making a new data set 

```python
new_df=df[(df['cgpa']<8.808933625397168) & (df['cgpa']>5.113546374602832)]
plt.figure(figsize=(16,9))
plt.subplot(2,2,1)
plt.subplot(2,2,1)
sns.distplot(df['cgpa'])
plt.title('before trimming ')

plt.subplot(2,2,2)
sns.boxplot(df['cgpa'])
plt.title('before trimming ')

plt.subplot(2,2,3)
sns.distplot(new_df['cgpa'])
plt.title('after trimming ')

plt.subplot(2,2,4)
sns.boxplot(new_df['cgpa'])
plt.title('after trimming ')
```
