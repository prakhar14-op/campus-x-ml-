POLYNOMIAL REGRESSION

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt 

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
x = np.linspace(-20, 20, 200).reshape(-1, 1)  # Clean spread on x-axis
y = 0.8 * x**2 + 0.9 * x + 2 + np.random.randn(200, 1) * 10  # Slight noise
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=2)
```

```python
lr=LinearRegression()
lr.fit(x_train,y_train)
y_pred=lr.predict(x_train)
```

```python
plt.plot(x_train,y_pred,color='r')
plt.plot(x.ravel(),y.ravel(),'b.')
```

applying polynomial linear regression 

```python
poly=PolynomialFeatures(degree=2)
x_train_trans=poly.fit_transform(x_train)
x_test_trans=poly.transform(x_test)
# print(x_train[0])
# print(x_train_trans[0])
```

```python
lr.fit(x_train_trans,y_train)
y_pred=lr.predict(x_test_trans)
r2_score(y_test,y_pred)
```

```python
plt.plot(x_train,lr.predict(x_train_trans),'r',linewidth=2,label='prediction')
plt.plot(x_train,y_train,'b.',label='training points')
plt.plot(x_test,y_test,'g.',label='testing points')
plt.xlabel('x')
plt.plot('y')
plt.legend()
plt.show()
```
